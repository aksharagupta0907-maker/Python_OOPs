                        Python OOPs Questions


1. What is Object-Oriented Programming (OOP)?
- OOP is a programming paradigm that organizes code into objects, which contain both data (attributes) and behavior (methods). It helps model real-world entities and improves code reusability, scalability, and maintainability.

2. What is a class in OOP?
- A class is a blueprint or template for creating objects. It defines the attributes and methods that the objects created from it will have.

3. What is an object in OOP?
- An object is an instance of a class. It represents a real-world entity and can access the class’s attributes and methods.

4. What is the difference between abstraction and encapsulation?
- Abstraction: Hides implementation details and shows only essential features (focuses on what an object does).

- Encapsulation: Bundles data and methods together and restricts direct access to data (focuses on how data is protected).

5. What are dunder methods in Python?
- Dunder methods (short for "double underscore methods"), also known as magic methods, are special methods in Python that have double underscores before and after their names (e.g., __init__, __str__). They are automatically invoked by Python in specific situations, such as object initialization or string representation.

6. Explain the concept of inheritance in OOP?
- Inheritance allows a class (child) to acquire properties and methods of another class (parent), promoting code reuse.

7. What is polymorphism in OOP?
- Polymorphism is the ability of an object to take on many forms. In OOP, it allows methods to do different things depending on the object it is acting upon, typically achieved through method overriding or method overloading (though Python handles the latter differently).

8. How is encapsulation achieved in Python?
- Encapsulation is achieved using:

  - Public attributes
  - Protected attributes (_variable)
  - Private attributes (__variable)
      and controlled access via methods or properties.

9. What is a constructor in Python?
- A constructor is a special method __init__() that is automatically called when an object is created. It initializes object attributes.

10. What are class and static methods in Python?
- Class method (@classmethod): Works with the class itself, not instances.
- Static method (@staticmethod): Does not access class or instance data; behaves like a regular function inside a class.

11. What is method overloading in Python?
- Python doesn’t support traditional method overloading. It’s simulated using default arguments or *args and **kwargs.

12. What is method overriding in OOP?
- Method overriding occurs when a child class provides its own implementation of a method already defined in the parent class.

13. What is a property decorator in Python?
- The @property decorator allows a method to be accessed like an attribute, enabling controlled access to instance variables.

14. Why is polymorphism important in OOP?
- Polymorphism improves flexibility, code reuse, and scalability by allowing objects of different classes to be treated uniformly.

15. What is an abstract class in Python?
-  An abstract class is a class that cannot be instantiated on its own but serves as a blueprint for other classes. They are used to define a common interface for subclasses and often contain one or more abstract methods (methods declared but not implemented in the abstract class). Python's abc module is used to create them.

16. What are the advantages of OOP?
- The advantages of OOP include:
  - Modularity : Objects create self-contained modules.
  - Reusability : Inheritance allows code reuse.
  - Maintainability : Easier to manage and debug code.
  - Flexibility : Polymorphism allows for flexible designs.
  - Encapsulation : Protects data integrity by hiding implementation details.

17. what is an the difference between a class variable and an instance variable?
-  Class variables are shared by all instances of a class and are defined directly within the class body.
- Instance variables are unique to each instance (object) of the class and are defined within the __init__ method using self.

18. What is multiple inheritance in Python?
- Multiple inheritance is a feature where a class can inherit from more than one parent class. This allows the inheriting class to access attributes and methods from all its parent classes, though it can introduce complexity with the Method Resolution Order (MRO).

19. Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in Python?
- The __str__ method is used to define the "informal" or user-friendly string representation of an object, typically used by print(). The __repr__ method defines the "official" or developer-focused string representation, which aims to be unambiguous and, if possible, a valid Python expression to recreate the object.

20. What is the significance of the ‘super()’ function in Python?
- The super() function is used to access methods and attributes of a parent or sibling class within an inheritance hierarchy. It is commonly used in the subclass's __init__ method to call the parent class's constructor, ensuring proper initialization of inherited components.

21. What is the significance of the __del__ method in Python?
- The __del__ method is a destructor in Python. It is called when an instance of a class is about to be destroyed (i.e., when its reference count reaches zero). It can be used to perform cleanup activities, such as closing file connections or releasing external resources, before the object is garbage collected.

22. What is the difference between @staticmethod and @classmethod in Python?
- A @staticmethod does not take a reference to the instance or the class as its first argument; it is just a function logically grouped within the class. A @classmethod takes the class itself (cls) as its first argument, allowing it to modify class state or create class-specific instances (alternative constructors).

23. How does polymorphism work in Python with inheritance?
- Polymorphism with inheritance works in Python because Python is dynamically typed. It doesn't care about an object's type, only that it supports the method being called (duck typing). A subclass can override a parent method, and code written to use the parent method will seamlessly use the subclass's version when passed an instance of the subclass.

24. What is method chaining in Python OOP?
- Method chaining is a programming technique where multiple methods are called sequentially on the same object, with each method returning the object itself to allow for the next call in the chain.

This allows for more concise and readable code, as operations can be performed in a single line. A common example is with string methods like
str.replace().split().lower().

25. What is the purpose of the __call__ method in Python?
- The __call__ method in Python allows an instance of a class to be called as if it were a function.  

When an object is called using parentheses, the __call__ method of its class is automatically executed. This is useful for creating objects that act like functions, such as decorators, or for objects that need to maintain state while being callable.

1. Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog that overrides the speak() method to print "Bark!".

```python
class Animal:
    def speak(self):
        print("Generic message")

class Dog(Animal):
    def speak(self):
        print("Bark!")

# Example usage:
animal = Animal()
animal.speak()

dog = Dog()
dog.speak()
```

2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return math.pi * self.radius**2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width

# Example usage:
circle = Circle(5)
print(f"Circle area: {circle.area()}")

rectangle = Rectangle(4, 6)
print(f"Rectangle area: {rectangle.area()}")
```

3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.

```python
class Vehicle:
    def __init__(self, type):
        self.type = type

class Car(Vehicle):
    def __init__(self, type, brand):
        super().__init__(type)
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, type, brand, battery_capacity):
        super().__init__(type, brand)
        self.battery_capacity = battery_capacity

# Example usage:
my_car = ElectricCar("Sedan", "Tesla", "75 kWh")
print(f"My car is a {my_car.brand} {my_car.type} with a {my_car.battery_capacity} battery.")
```

5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.

```python
class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.__balance}")
        else:
            print("Insufficient funds or invalid amount.")

    def get_balance(self):
        return self.__balance

# Example usage:
account = BankAccount(100)
account.deposit(50)
account.withdraw(25)
print(f"Current balance: ${account.get_balance()}")
# Trying to access the private attribute directly will result in an error:
# print(account.__balance)
```

6.  Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().

```python
class Instrument:
    def play(self):
        print("Instrument playing generic sound.")

class Guitar(Instrument):
    def play(self):
        print("Guitar plays a melody.")

class Piano(Instrument):
    def play(self):
        print("Piano plays a harmony.")

# Example usage:
def play_instrument(instrument):
    instrument.play()

guitar = Guitar()
piano = Piano()

play_instrument(guitar)
play_instrument(piano)
```

7.  Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.

```python
class MathOperations:
    class_name = "Math Operations"

    @classmethod
    def add_numbers(cls, a, b):
        print(f"Using {cls.class_name}:")
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

# Example usage:
sum_result = MathOperations.add_numbers(10, 5)
print(f"Sum: {sum_result}")

diff_result = MathOperations.subtract_numbers(10, 5)
print(f"Difference: {diff_result}")
```

8.  Implement a class Person with a class method to count the total number of persons created.

10.  Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

# Example usage:
v1 = Vector(2, 3)
v2 = Vector(4, 1)
v3 = v1 + v2
print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}")
print(f"Sum of vectors: {v3}")
```

12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.

```python
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)
```

14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.

```python
class Employee:
    def calculate_salary(self, hours_worked, hourly_rate):
        return hours_worked * hourly_rate

class Manager(Employee):
    def calculate_salary(self, hours_worked, hourly_rate, bonus=0):
        base_salary = super().calculate_salary(hours_worked, hourly_rate)
        return base_salary + bonus
```

16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        return "Moo"

class Sheep(Animal):
    def sound(self):
        return "Baa"
```
